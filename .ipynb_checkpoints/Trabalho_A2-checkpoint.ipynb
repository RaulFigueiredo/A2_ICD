# Trabalho de Análise Exploratória de Dados com Python


Importando as bibliotecas necessárias:

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importando a base de dados:
 Essa base de dados possui informações sobre os top 10 filmes que mais lucraram por ano, desde 1975 até 2018. A base contém o gênero de cada filme, a nota do IMDB, a duração, a colocação do filme em seu ano, o estúdio, a recomendação de idade, título, total de lucro e ano de lançamento. 

```python
filmes = pd.read_csv('dados/bilheteria.csv')
```

# Conhecendo e limpando a base


Essa é a base de dados antes de qualquer alteração

```python
filmes.head()
```

```python
filmes.tail(4)
```

Verificando valores faltantes:

```python
filmes.isnull().sum()
```

Retirando as coluna genero 2 e genero 3 que não usaremos 

```python
filmes.drop(columns={'Genre_3','Genre_2'}, inplace=True)
#Observamos que vários filmes não possuem subgêneros, 
#por isso retiramos as colunas Genre_3 e Genre_2 da base, já que não serão utilizadas em nossa análise.
#inplace=True faz essa modificação ser aplicada e salva.
```

Reordenamos as colunas para facilitar a leitura da base de dados.

```python
filmes=filmes[['title','Main_Genre','studio','imdb_rating','worldwide_gross','length','rating','rank_in_year','year']]
filmes
```

```python
filmes.info()
```

Nesse momento ordenamos a base de dados de acordo com o lucro apenas por curiosidade, nenhuma análise foi feita 

```python
filmes.sort_values('worldwide_gross', ascending=False)
#Ordena a base de dados de acordo com o valores de "worldwide_gross", de maneira decrescente.
```

# Análise exploratória

## Quais são os filmes com as maiores notas?

```python
filmes[['title','Main_Genre', 'imdb_rating','year','worldwide_gross']].sort_values('imdb_rating',ascending=False)
#Ordenando a base de acordo com a nota imdb de forma descrescente
```

Essa tabela é muito interessante, ordena os filmes de acordo com a sua nota de IMDB e fica muito facil de se ver ou analisar.

```python
top_10 = filmes[['title','imdb_rating']].sort_values('imdb_rating', ascending=False)

name = [name for name in top_10.iloc[:, 0]] #extraindo os nomes

plt.subplots(figsize=(12,6))
plt.barh(name[:10], top_10['imdb_rating'][:10], color='royalblue', label='ratings') 
plt.xlabel('Nota IMDB',
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})
plt.title('Top 10 filmes em relação à nota', 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 22},
          loc='left')
#gráfico de barras com os 10 primeiros filmes de acordo com a nota
```

A única informação que conseguimos com esse simples gráfico é que as notas de imdb dos top 10 filmes estão bem próximas, não há outliers.

```python
#apenas uma curiosidade aleatoria
flt = ((filmes.imdb_rating>8) & (filmes.year>2013))
#Filtrando apenas os filmes com nota imdb maior que 8, lançados após 2013.
filmes[flt][['title','Main_Genre','imdb_rating','year']]
#Exibindo apenas as colunas desejadas
```

```python
#apenas uma curiosidade aleatoria
flt = ((filmes['Main_Genre']=='Action') & (filmes.imdb_rating>8))
#Filtrando filmes de ação com nota imdb maior que 8
filmes[flt][['title','imdb_rating','year']]
#Exibindo apenas as colunas desejadas
```

## Qual o melhor estúdio cinematográfico?

### De acordo com o lucro:



```python
lucro_por_studio = filmes.groupby('studio').sum()
lucro_por_studio = lucro_por_studio[['worldwide_gross']]
lucro_por_studio = lucro_por_studio.sort_values('worldwide_gross')
#Tabela com o estúdio como index e a soma do 'worldwide_gross' de cada estúdio como valor
lucro_por_studio.plot(kind='barh',figsize=(14,9), color="turquoise")
#Gráfico de barra horizontal 14 por 9.
plt.xlabel('Lucro total, em bilhões de dolares',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.ylabel('Estúdios',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.title('Lucro total por estúdio', 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 22},
          loc='left')
#Uma fonte, para título, eixo x e eixo y, bonitinha. Essa forma fixa de legenda foi usada em quase todos os gráficos.

plt.legend().remove()
#Remove a legenda que não era útil.
```

A Warner Bross e a Walt Disney Pictures foram as companhias com maior lucro total e, portanto, pode-se dizer que as mais bem sucedidas. No entanto, a grande diferença de lucro dessas duas empresas em relação às outras pode ter sido causada por algum outlier ou até mesmo pela grande quantidade de filmes já produzidos por elas.

## De acordo com a nota:

```python
nota_por_studio = filmes.groupby('studio').mean()
nota_por_studio = nota_por_studio[['imdb_rating']]
nota_por_studio = nota_por_studio.sort_values('imdb_rating')
nota_por_studio.plot(kind='barh',figsize=(14,9), color="royalblue")
plt.xlabel('Nota média IMDB',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.ylabel('Estúdios',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.title('Nota IMDB por estúdio', 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 22},
          loc='left')

plt.legend().remove()
```

Ao analisar a média da nota IMDB de cada estúdio, a Pixar ganha o pódio enquanto a Disney e a Warner Bross decaem para o quarto e quinto lugar, respectivamente, mostrando que nem sempre grandes sucessos de bilheteria são considerados grandes sucessos pela crítica.

Concluímos então que a pergunta do “melhor estúdio” é relativa, depende se analisarmos o IMDB ou o lucro. Poderíamos distribuir pesos para IMDB e para o lucros e chegar a alguma conclusão mais precisa, mas os pesos seriam arbitrários, assim chegaríamos a um resultado não necessariamente correto.

## O faturamento acompanha a nota IMDB?

### Lucro por filme
- Antes de fazermos essa análise é interessante conhecermos melhor a coluna do lucro dos filmes 

```python
round((filmes.worldwide_gross.describe()/1000000),2)
```

Assim obtemos a média, a variação e os principais quartis do lucro por filme para facilitar a análise.

```python
#Esse gráfico não exigiu nenhuma soma, por isso o groupby não foi utilizado
imdb_bilheteria= filmes[['imdb_rating','worldwide_gross']]

plt.subplots(figsize=(12,8))
plt.plot([imdb_bilheteria.imdb_rating],[(imdb_bilheteria.worldwide_gross/1000000)],'co')
#Gráfico normal do matplotlib com pontos  
#worldwide_gross foi dividido por 1000000 para facilitar a visualização
plt.xlabel('Nota no IMDB',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 14})

plt.ylabel('Lucro total, em milhões de dolares',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 14})

plt.title('Relação entre a nota e o lucro em bilheteria', 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 20},
          loc='left')
plt.show()
```

```python

imdb_bilheteria= filmes[['imdb_rating','worldwide_gross']]

plt.subplots(figsize=(12,8))
plt.axis([0, 10, 0,2000])
#Escala fixa
#Unico detalhe diferente do outro gráfico
plt.plot([imdb_bilheteria.imdb_rating],[(imdb_bilheteria.worldwide_gross/1000000)],'co')
plt.xlabel('Nota no IMDB',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 14})

plt.ylabel('Ganho em bilheteria, em milhões de dolares',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 14})

plt.title('Relação entre a nota e o lucro em bilheteria', 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 20},
          loc='left')
plt.show()
```

- Fixando a escala eu retirei um outlier e foi possível visualizar melhor a relação entre a nota imdb e o ganho na bilheteria. É possivel ver que há, relativamente, poucos filmes que entraram no top 10 com nota menor que 5, e entre as notas 6 e 8 há uma aglomeração de pontos.
- Não consegui demonstrar a minha ideia de total correlação entre o lucro do filme e sua nota, para isso seria preciso que os pontos formassem uma 'linha' crescente, mostrando que quanto maior a nota, maior o ganho na bilheteria. 
- Assim concluímos que o faturamento não, necessariamente, acompanha o IMDB.

## Quais gêneros fazem maior sucesso?

Calculando o número total de filmes em cada gênero:

```python
filmes['Main_Genre'].value_counts()
```

### Lucro total de cada gênero:

```python
#A soma do 'worldwide_gross' por gênero era necessária, então o groupby foi utilizado.
lucro_por_genero = filmes.groupby('Main_Genre').sum()
lucro_por_genero=lucro_por_genero[['worldwide_gross']]
lucro_por_genero=lucro_por_genero.sort_values('worldwide_gross')

lucro_por_genero.plot(kind='barh',figsize=(14,9),color='turquoise')

plt.xlabel('Ganho em bilheteria, em dezenas de bilhões de dolares',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.ylabel('Gêneros',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.title('Lucro total por gênero', 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 22},
          loc='left')
plt.legend().remove()
```

Fantasia possui os filmes mais lucrativos, quase 10 bilhões de dolares à frente do próximo gênero, ficção científica que esta logo em segundo lugar.

### Nota IMDB de cada gênero

```python
imdb_por_genero = filmes.groupby('Main_Genre').mean()
imdb_por_genero=imdb_por_genero[['imdb_rating']]
imdb_por_genero=imdb_por_genero.sort_values('imdb_rating')

imdb_por_genero.plot(kind='barh',figsize=(14,8),color='RoyalBlue')

plt.xlabel('Média das notas do IMDB',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.ylabel('Gênero',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.title('Nota média de IMDB por gênero', 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 22},
          loc='left')
plt.legend().remove()
plt.axis([0, 10,-0.5,15.5])
plt.grid()
#O grid foi utilizado para deixar mais claro que a nota imdb de todos os gêneros estão entre 6 e 8.
```

Já em relação à opinião dos críticos do IMDB, Guerra e História são os gêneros com os melhores filmes. Mas o intervalo em que as notas varia é pequeno, entre 6 e 8.

Então fica claro que o gênero de maior sucesso de vendas é "Fantasia" com melhor media de IMDB é "Guerra".

## Análise temporal 

### Variação do valor de bilheteria ao longo dos anos


```python
lucro_por_ano = filmes.groupby('year').sum()
lucro_por_ano=lucro_por_ano[['worldwide_gross']]
lucro_por_ano=lucro_por_ano.sort_values('year')
lucro_por_ano
lucro_por_ano.plot(kind='line',figsize=(14,8),linewidth=5.0,color='turquoise')
plt.plot([lucro_por_ano.index],[(lucro_por_ano.worldwide_gross)],'bo')
#Além do gráfico de linhas plotado pelo pandas, fiz uma representação de pontos azuis usando o matplotlib.
#Os pontos marcam cada ano e ajudam a visualizar melhor o gráfico.

plt.xlabel('Ano',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.ylabel('Lucro em bilhões de dolares',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.title('Lucro ao decorrer dos anos', 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 22},
                    loc='left')
plt.legend().remove()
plt.xticks([1975,1980,1985,1990,1995,2000,2005,2010,2015,2018])#defini os pontos que vão aparecer no eixo X
plt.grid()
```

Existe um aumento geral do lucro ao longo dos anos, desde o final da década de 80 até 2012, com algumas excessões e decaimentos. 

Os maiores picos são em 2009 e 2012, provavelmente por causa dos lançamentos de Avatar e Os Vingadores.

Após 2014 o lucro sofreu uma grande queda.

### Variação da nota IMDB ao longo dos anos

```python
imdb_por_ano = filmes.groupby('year').mean()
imdb_por_ano=imdb_por_ano[['imdb_rating']]
imdb_por_ano=imdb_por_ano.sort_values('year')
imdb_por_ano
imdb_por_ano.plot(kind='line',figsize=(14,8),linewidth=5.0, color='turquoise')
plt.plot([imdb_por_ano.index],[(imdb_por_ano.imdb_rating)],'bo')

plt.xlabel('Anos',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.ylabel('Bilheteria em bilhões',          
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})

plt.title('Média da nota IMDB ao decorrer dos anos', 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 22},
                    loc='left')
plt.legend().remove()
plt.xticks([1975,1980,1985,1990,1995,2000,2005,2010,2015,2018])
plt.grid()
```

Vemos que as variáveis parecem não estar relacionadas, mas temos alguns picos que representam os anos que mais produziram boas obras, sendo os maiores 2013 e 2018.

## Qual a influencia da duração na qualidade de um filme?


### Duração dos filmes mais lucrativos

```python
top_10 = filmes[['title', 'length', 'worldwide_gross']].sort_values('worldwide_gross', ascending=False)
name = [name for name in top_10.iloc[:, 0]] #extraindo os nomes

plt.subplots(figsize=(12,8))
plt.barh(name[:10], top_10['length'][:10], color='turquoise', label='length(hours)')
plt.title('Duração do TOP 10 filmes (lucro)',
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 22},
                    loc='left')
plt.xlabel('Duração (minutos)',
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})
```

A duração da maioria dos filmes que mais lucraram se concentra no intervalo entre um grande intervalo, de 125 a 175 minutos. Frozen foge desse intervalo sendo o filme mais curto (cerca de 100 minutos apenas), e Titanic e Senhor dos Aneis são os mais longos, quase 200 minutos.

### Duração dos filmes com maior nota

```python
top_10 = filmes[['title', 'length', 'imdb_rating']].sort_values('imdb_rating', ascending=False)
name = [name for name in top_10.iloc[:, 0]] #extraindo os nomes

plt.subplots(figsize=(12,8))
plt.barh(name[:10], top_10['length'][:10], color='royalblue', label='length(hours)')
plt.title('Duração do TOP 10 filmes (nota)',
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 22},
                    loc='left')
plt.xlabel('Duração (minutos)',
           fontdict={'family': 'serif', 
                    'color' : 'black',
                    'weight': 'bold',
                    'size': 16})
```

Ao realizar a análise comparando os filmes melhor avaliados percebemos que o intervalo se mantém entre 125 e 175 minutos, sendo as exceções os dois filmes mais longos, Senhor dos Aneis: O Retorno do Rei e Schindler's List, respectivamente.

Desta forma é possível ver que o tempo dos filmes não tem ligação com a qualidade do filme.  

Para uma analise mais precisa do tempo dos filmes, é interessante olharmos as estatísticas gerais do tempo dos filmes  

```python
round((filmes.length.describe()),2)
```

Obtemos a média, a variação e os principais quartis da duração dos filmes.

Concluindo, portanto, que a duração do filme não é um fator decisivo, nem para seu lucro, nem para sua nota IMDB.

# Conclusões

Tomando como ponto de partida as questões levantadas e escritas na planilha, questões essas que foram feitas antes da análise dos dados, podemos chegar as seguintes conclusões:

-  Vimos claramente a duração não é um fator decisivo para a qualidade de um filme, analisando os top 10 nota IMDB e top 10 Lucro enxergamos que  a duração do filme beira o aleatório.
-  A ideia de melhor estúdio parece com a do melhor gênero. Se levarmos em conta o lucro, Warner Bros e Disney saem disparados, porém se olharmos a média do IMDB vemos a Pixar em primeiro lugar com uma distância interessante do segundo lugar, e vale ressaltar que a Pixar ficou em decimo lugar dos estúdios que mais tiveram lucro. Então seria preciso criar pesos arbitrários para decidir qual o melhor estúdio com clareza. 
- Concluímos também que a nota do IMDB não é um fator decisivo para o lucro em bilheteria, o que realmente me surpreendeu, particularmente achei que teria uma relação clara entre os dois fatores.
